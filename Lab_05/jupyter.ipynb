8 Queen Problem using Hill Climbing Algorith

```python
import random

# ── 1. State Representation ──────────────────
# board = list of length 8
# board[col] = row position of the queen in that column


def generate_random_state():
    """Generate a random initial board configuration."""
    return [random.randint(0, 7) for _ in range(8)]
```

```python
# ── 2. Heuristic Function ────────────────────
def heuristic(board):
    """
    Count the number of attacking queen pairs.
    h(n) = 0  →  solution found.
    """
    attacks = 0
    n = len(board)
    for i in range(n):
        for j in range(i + 1, n):
            # Same row
            if board[i] == board[j]:
                attacks += 1
            # Same diagonal
            if abs(board[i] - board[j]) == abs(i - j):
                attacks += 1
    return attacks
```

```python
# ── 3. Neighbour Generation ──────────────────
def get_neighbours(board):
    """
    Generate all neighbours by moving each queen
    to every other row in its column.
    Total neighbours = 8 columns × 7 moves = 56.
    """
    neighbours = []
    n = len(board)
    for col in range(n):
        for row in range(n):
            if row != board[col]:           # skip current position
                neighbour = board[:]        # copy current state
                neighbour[col] = row        # move queen
                neighbours.append(neighbour)
    return neighbours
```

```python
# ── 4. Hill Climbing ─────────────────────────
def hill_climbing():
    """
    Steepest-ascent (lowest h) Hill Climbing.
    Returns (solution_or_local_optimum, h_value, steps, success).
    """
    current = generate_random_state()
    current_h = heuristic(current)
    steps = 0

    print("=" * 50)
    print("  8-QUEENS — HILL CLIMBING")
    print("=" * 50)
    print(f"\nInitial state : {current}")
    print(f"Initial h(n)  : {current_h}  (attacking pairs)\n")

    while True:
        neighbours = get_neighbours(current)

        # Find the neighbour with the lowest heuristic
        best_neighbour = None
        best_h = current_h

        for nb in neighbours:
            h = heuristic(nb)
            if h < best_h:
                best_h = h
                best_neighbour = nb

        steps += 1

        if best_neighbour is None:
            # No improvement found → local optimum or solution
            if current_h == 0:
                print(f"  Solution found in {steps} step(s)!")
                return current, current_h, steps, True
            else:
                print(f"  Local optimum reached after {steps} step(s).")
                print(f"   Stuck at h(n) = {current_h}")
                return current, current_h, steps, False

        # Move to the best neighbour
        current = best_neighbour
        current_h = best_h
        print(f"Step {steps:>3} | state: {current} | h(n) = {current_h}")

        if current_h == 0:
            print(f"\n  Solution found in {steps} step(s)!")
            return current, current_h, steps, True
```

```python
# ── 5. Board Visualiser ──────────────────────
def print_board(board):
    """Pretty-print the chessboard."""
    n = len(board)
    print("\n  " + "─" * (n * 4 + 1))
    for row in range(n):
        line = f"{row} │"
        for col in range(n):
            line += " Q │" if board[col] == row else " . │"
        print(line)
    print("  " + "─" * (n * 4 + 1))
    print("    " + "   ".join(str(c) for c in range(n)))
    print("    (columns)\n")
```

```python
# ── 6. Run with Restarts ─────────────────────
def run_with_restarts(max_restarts=100):
    """
    Restart from a new random state whenever a
    local optimum is hit, until a solution is found.
    """
    for attempt in range(1, max_restarts + 1):
        print(f"\n{'━'*50}")
        print(f"  Attempt {attempt}")
        print(f"{'━'*50}")

        state, h, steps, success = hill_climbing()

        if success:
            print(f"\nFinal board state : {state}")
            print_board(state)
            print(f"Total attempts    : {attempt}")
            print(f"Steps this run    : {steps}")
            return state

    print("\n❌  Could not find a solution within the restart limit.")
    return None
```

```python
# ── 7. Entry Point ───────────────────────────
if __name__ == "__main__":
    random.seed()          # fresh seed each run
    solution = run_with_restarts(max_restarts=100)
```
